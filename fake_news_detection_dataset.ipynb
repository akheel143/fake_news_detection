```
import pandas as pd
import matplotlib.pyplot as plt
import re
```

```
data=pd.read_csv('fake_news dataset.csv')
data=data.drop(['URLs','Body'],axis=1)
```

```
data.isnull().sum()
```

```
nltk.download('wordnet')
```

```
import re 
import nltk
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
leming=WordNetLemmatizer()
corpus=[]
for i in range(len(data['Headline'])):
    rewiew=re.sub('[^a-zA-Z]','',data['Headline'][i])
    rewiew=rewiew.lower()
    rewiew=rewiew.split()
    rewiew=[leming.lemmatize(word) for word in rewiew if not word  in set(stopwords.words('english'))]
    rewiew=''.join(rewiew)
    corpus.append(rewiew)
print(corpus)
```

```
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(corpus,data['Label'],test_size=0.2)
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
cv=TfidfVectorizer(lowercase=False)
train=cv.fit_transform(x_train).toarray()
```

```
test=cv.transform(x_test).toarray()
```

```
from sklearn.svm import SVC
reg=SVC()
reg.fit(train,y_train)
```

```
reg.score(train,y_train)
```

```
y_pred=reg.predict(test)
```

```
y_pred.shape
```

```
y_test
```

```
from sklearn.metrics import confusion_matrix,classification_report
print(classification_report(y_test,y_pred))
```
